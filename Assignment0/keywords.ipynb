```python
txt_names = ["trade-wars-news1.txt","trade-wars-news2.txt","trade-wars-news3.txt","trade-wars-news4.txt","trade-wars-news5.txt"]

all_txts = []
for i in txt_names:
   f=open (i,"r+",encoding="utf8")
   txt= f.read()
   all_txts.append(txt)

all_words = "".join(all_txts)
all_words
```

```python
all_words = str.lower(all_words)
all_words
```

```python
all_words = str.split(all_words)
```

```python
all_words
```

```python
freq = []
for x in all_words:
    times = all_words.count(x)
    freq.append(times)
```

```python
freq
```

```python
len(freq)
```

```python
len(all_words)
```

```python
freq_dict = dict(zip(all_words,freq))
freq_dict
```

```python
len(freq_dict)
```

```python
frequency = sorted(freq_dict.items(), key=lambda d:d[1],reverse=True)
frequency
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
for x in stop_words:
    for i in frequency:
        if i[0]==x:
           frequency.remove(i)
frequency
```

```python
top_15_keywords = frequency[:15]
top_15_keywords
```

```python
import csv
with open('keywordsfrequency.csv','w',newline='',encoding='utf8') as f:
    writer = csv.writer(f)
    header = ['keyword','frequence']
    writer.writerow(header)
    writer.writerows(top_15_keywords)
```
